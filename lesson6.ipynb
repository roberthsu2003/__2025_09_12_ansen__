```python
#for...in迴圈大部份應用在-"明確知道執行的次數"
total = 0
for i in range(1,11):
  total += i
print(total)
```

```python
# while不知道執行的次數
import random

while random.randint(1, 10) != 5:
  print("亂數不等於5")
print("亂數是5,結束應用程式")
```

```python
# 小明想要存錢買一輛機車，機車每輛30000元，他將每月存的錢輸入，當存款足夠買機車時，就顯示提示訊息告知。

deposit = 0

while deposit < 30000:
  value = int(input("本月份存款的金額:"))
  deposit += value
print(f"恭喜!已經存夠了,存了{deposit}")
```
