```python
# 巢狀迴圈
for i in range(1,10):
  #print("i=",i)
  for j in range(1,10):
    print(i,"*",j,"=",i*j,end="  ")
  print("\n=============")
```

```python
total = 0
for i in range(1,101):
  if i % 2 == 0:
    total += i

print("1到100偶數加總是",total)
```
